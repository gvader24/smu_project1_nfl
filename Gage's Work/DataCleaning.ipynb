```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("../Resources/spreadspoke_scores.csv", encoding="unicode_escape")
mask = (df.schedule_season >= 2000) & (df.schedule_season < 2020) & (df.schedule_playoff != True) & (df.stadium_neutral != True)
df = df.loc[mask].reset_index(drop=True)
df.head()
```

```python
df_games = df

df_games['team_home'] = df['team_home'].map(id_dict)
df_games['team_away'] = df['team_away'].map(id_dict)

df_games
```

```python
df_games.info()
```
